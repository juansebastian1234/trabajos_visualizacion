#Visualizacion 1: Artistas mas vendidos desde el año 1969



visualizacion 1.PNG

##¿Qué?
Esta es una visualización de los artistas más vendidos desde el año 1969 a partir del 4 tercio del año. 
##¿Por qué?
Al ser una visualización dinámica, donde se muestra el top 10 de los artistas más vendidos, el usuario debe estar observando las 10 líneas donde se muestran los artistas o grupos con más ventas.
##¿Comó?
Los modismos que uso esta visualizacion, es de barras en forma horizontal, donde cada una de las 10 barras presentes representaban al top 10 artistas mas vendidos en ese momento, todas las barras tienen colores, algunas de estas tienen los mismos colores, pero no hay un significado de cada color.
##Marcas y canales
En este punto se evidencia que la visualización utiliza como marca a las líneas, usando como canal de posición el espacio en posición horizontal, al ser una visualización dinámica, las líneas tienen un diferente tamaño a medida que se va reproduciendo el video.
##Reglas
Esta visualización está cumpliendo algunas reglas generales, cumple la regla del overview, donde la información general es fácilmente plasmada con una gran facilidad para su visualización, también otro punto a favor de la visualización es que opta por el camino fácil para mostrar los datos, ya que al ser un plano 2D las líneas horizontales cumplen un buen trabajo ya que para este caso sería innecesario un gráfico 3D, otra regla que cumple es la velocidad de respuesta, esto es bueno ya que al tener una rápida capacidad, no genera en el público un rechazo. Hay otras reglas que no cumple esta visualización como la memoria ocular, y no la cumple ya que toda la visualización es dinámica, es decir, todas las barras están en pleno movimiento, por lo que es más complicado observar todos los datos.
##Mejoras
Como se había mencionado antes, una mejora para esta visualización es la parte del dinamismo en la visualización, al estar todas la barras en movimiento se está perdiendo información, para poder mejorar este apartado, una opción podría ser que solo una o dos de las barras se mueva, ya que al tan solo moverse una o dos, es más fácil que se retenga la información, otra cosa a mejorar es en la paleta de colores, ya que durante toda la visualización muchas barras tienen el mismo o un color muy similar, por lo que no se logra diferenciar en varios casos.


#Visualización 2: Proyecto de salud de Trump

visualizacion 2.PNG

##¿Qué?
Esta es una visualización en donde se muestra el proyecto de ley médica del ese entonces presidente de los Estados Unidos Donald Trump.
##¿Por qué?
Esta es una visualización estática, en donde los usuarios pueden observar por condados como es la favorabilidad de este proyecto de ley con respecto a los votantes del país. 
##¿Cómo?
El modismo utilizado en esta visualizacion es por areas, donde estas represetan a los condados por cada estado en Estados Unidos, los condados estan con diferentes colores, dodne cada uno refleja la aceptacion del proyecto de ley en los votantes de Trump.
## Marcas y Canales
Esta visualización está usando una marca de área, donde además está usando como canal también áreas, estas áreas serían los condados que se observan en el mapa, donde existen espacios más grandes a comparación de otros.

##Reglas
En esta visualización se evidencia que no se cumple una de las reglas más características y es la regla del overview, esto se puede justificar ya que a simple vista el mapa nos ofrece la visión de los votantes de Trump a su proyecto de salud, solo que el mapa al ser tan denso y estar fraccionado en diferentes condados, hace que su análisis sea más complicado, otra regla que también no se está ejecutando adecuadamente es la regla de la funcionalidad y luego la forma, esto se debe a que la forma a como fue creada esta visualización no funciona,  ya que no es práctica para un análisis fácil.
##Mejoras
En primera si es obligatorio usar el mapa, seria cambiar la paleta de colores, ya que al ser colores muy similares, genera un conflicto para hacer más prácticos los análisis, por otra parte, también se podría mejorar la parte del enfoque del mapa, ya que al ser un mapa tan grande que incluye tantos condados, se podría fraccionar el mapa para un análisis más eficaz y sencillo.
